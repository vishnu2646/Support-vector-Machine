## Data Preprocessing

```python
data.isnull().sum()
```

```python
# check how many values are null in GENDER COLUMN
data.loc[data['Gender'].isnull()==True, 'Gender'] = 'Male'
```

```python
data.loc[data['Married'].isnull()==True, 'Married'] = 'Yes'
```

```python
data.loc[data['Dependents'].isnull()==True, 'Dependents'] = '3+'
```

```python
data.loc[data['Self_Employed'].isnull()==True, 'Self_Employed'] = 'Yes'
```

```python
data.loc[data['LoanAmount'].isnull()==True, 'LoanAmount'] = np.median(data.LoanAmount.dropna())
```

```python
data.Loan_Amount_Term.value_counts()
```

```python
data.loc[data['Loan_Amount_Term'].isnull()==True, 'Loan_Amount_Term'] = 360
```

```python
data.Credit_History.value_counts()
```

```python
data.loc[data['Credit_History'].isnull()==True, 'Credit_History'] = 0
```

```python
data.isnull().sum()
```

## Encoding

```python
data['Gender'] = data.Gender.replace({'Female': 0, 'Male': 1})
```

```python
data
```

```python
## Oneheart encoding
data['Married'] = data.Married.replace({'Yes': 1, 'No': 0})
```

```python
data['Education'] = data.Education.replace({'Graduate': 1, 'Not Graduate': 0})
```

```python
data['Self_Employed'] = data.Self_Employed.replace({'No': 0, 'Yes': 1})
```

```python
data
```

```python
encoder = LabelEncoder()
test = encoder.fit_transform(data.Dependents)
```

```python
data['Dependents']=pd.DataFrame(test)
```

```python
data
```

```python
data2 = data.iloc[:,6:9]
data2
```

## Scaling

```python
scalar = StandardScaler()
data3 = scalar.fit_transform(data2)
```

```python
data[['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount']] = pd.DataFrame(data3)
```

```python
data
```

```python
data['Loan Status'] = data['Loan Status'].replace({'Y': 1, 'N': 0})
```

```python
data
```

## Spliting the data

```python
X = data.drop(['Loan Status', 'Loan_ID'], axis=1)
```

```python
y = data['Loan Status']
```

```python
# check for value_counts to y to check its is a balanced data or unbalanced dataset
y.value_counts()
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, random_state=54, test_size=0.2)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
smote = SMOTE()
x_train_sm, y_train_sm = smote.fit_resample(x_train, y_train)
```

```python
from collections import Counter

print(Counter(y_train))
print(Counter(y_train_sm))
```

## Model Implementation

```python
svm = SVC()
svm.fit(x_train_sm, y_train_sm)
```

```python
y_pred = svm.predict(x_test)
```

## Evaluation

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
accuracy_score(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

```python
confusion_matrix(y_test, y_pred)
```

## Hyperparameter tunning

```python
from sklearn.model_selection import GridSearchCV

param_grid = {'C': [0.1, 5, 10, 20, 30, 40, 50, 60, 70], 'gamma': [1, 0.1, 0.001, 0.0001], 'random_state': (list(range(1, 20)))}
model = SVC()
grid = GridSearchCV(model, param_grid, verbose=2, cv=5, scoring='f1')
grid.fit(X, y)
```
